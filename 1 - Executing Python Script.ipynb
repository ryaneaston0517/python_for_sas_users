## Simple Script Example

Below is a simple script illustrating a python for loop.  The set of numbers contained inside the square brackest [ ] make up the elements of a Python list.  In Python, a list is a data structure that holds an arbitrary collection of items.  The variable i is used as the index into the loop.  The variable product holds the integer value from the arithmetic assignment of product * i.  And finally, the print function displays the output.

```python
#First Python program using a for loop

numbers = [2, 4, 6, 8, 11]
product = 1
for i in numbers:
    product = product * i
print('The product is:', product)
```

## Line Continuation Symbol

Should you find you have a line of code needing to extend past the physical line (i.e., wrap), then use the backslash (\). This causes the Python interpreter to ignore the physical end-of-line termniator for the current line and continues scanning for the next end-of-line terminator.  The Python line continuation symbol is show below:

```python
y = 1 + \
    2
print(y)
```
